## 2. Chạy script (khuyến nghị: subprocess — tách process, ổn định)

Sửa đường dẫn checkpoint và `--backend` (`sit` = repo SiT gốc, `repa` = REPA).

````python
from pathlib import Path
import subprocess
import sys

WORKSPACE = Path("/workspace")
SCRIPT = WORKSPACE / "plot_sit_xl2_activation_layers.py"

# --- Sửa các biến này ---
CKPT = WORKSPACE / "SiT" / "pretrained_models" / "SiT-XL-2-256x256.pt"  # ví dụ
OUTDIR = WORKSPACE / "outputs" / "sit_activation_heatmap"
BACKEND = "sit"  # "sit" | "repa"

OUTDIR.mkdir(parents=True, exist_ok=True)

cmd = [
    sys.executable,
    str(SCRIPT),
    "--ckpt", str(CKPT),
    "--outdir", str(OUTDIR),
    "--backend", BACKEND,
    "--fast",           # bỏ "--fast" để chạy đủ sample (mặc định ~512)
    "--seed", "0",
]
# REPA ví dụ thêm:
# cmd += ["--repa-args-json", "/path/to/args.json", "--projector-embed-dims", "768"]

print(" ".join(cmd))
subprocess.run(cmd, check=True, cwd=str(WORKSPACE))
````

## 3. Xem ảnh PNG vừa lưu

````python
from IPython.display import Image, display
from pathlib import Path

# Dùng OUTDIR từ ô chạy script; nếu chưa chạy ô đó thì đổi đường dẫn dưới đây
try:
    _out = OUTDIR
except NameError:
    _out = Path("/workspace/outputs/sit_activation_heatmap")

png = _out / "activations_sit_xl2.png"
if png.is_file():
    display(Image(filename=str(png)))
else:
    print("Chưa có file:", png)
````
